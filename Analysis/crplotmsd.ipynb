# MSD plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
folder="../../bulk/production.bulk.cubic"
```

```python
a=np.load(folder+"/MSD.npy")
```

```python
time=a[1:,0]/1000
msd=a[1:,1]
logt=np.log10(time)
logm=np.log10(msd)
```

```python
plt.loglog(time,msd,label='data',alpha=0.8)

#slope,intercept=np.polyfit(logt[200:],logm[200:],1)
slope,intercept=np.polyfit(logt[20:],logm[20:],1)
fitted_y=slope* logt[20:]+intercept
plt.plot(time[20:],10**fitted_y, color="red", label = 'Fitted',linestyle='dashed')
plt.ylabel("MSD ($A^2$)",fontsize=16)
plt.text(200, 5.95, 'slope= %0.2f'%(slope), fontsize=12, color='green',ha='left', va='top')
plt.xlabel("Time(ps)",fontsize=16)
plt.legend()
plt.savefig(folder+"/msdloglog.png",dpi=1000)
plt.show()
```

```python
plt.plot(time,msd,label='data',alpha=0.8)

#slope,intercept=np.polyfit(logt[200:],logm[200:],1)
slope,intercept=np.polyfit(time[20:],msd[20:],1)
fitted_y=slope* time[20:]+intercept
plt.plot(time[20:],fitted_y, color="red", label = 'Fitted',linestyle='dashed')
plt.ylabel("MSD ($A^2$)",fontsize=16)
plt.text(600, 15, 'slope= %0.2f'%(slope), fontsize=12, color='green',ha='left', va='top')
D=(slope*100/6)*10**(-6)
plt.text(600, 10, 'D= %1.2f \u00D7 $10^{-7}$ $cm^2/s$'%(D*10**7), fontsize=12, color='green',ha='left', va='top')
plt.xlabel("Time(ps)",fontsize=16)
plt.legend()
plt.savefig(folder+"/msd.png",dpi=1000)
plt.show()
```
